```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
```

```python
driver = webdriver.Chrome()
url= "https://www.biostadt.com/products/adjuvants.aspx"
driver.get(url)


wait = WebDriverWait(driver, 10) # wait for up to 10 seconds for the page to load
wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))


content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")

div = soup.find("ul",{"class":"prod-list"})
# ul = div.find("ul")
links = div.find_all("li")


# imglist = []
productlinks = []
for i in links:
    a = i.find("a")
    link = a.get("href")
    productlinks.append(link)  
#     img = i.find("img").get("src")
#     imglist.append(img)

print(len(productlinks))


driver.quit()
```

```python
datas = []
for x in range(0, len(productlinks)):


#     if x == 3:
#         break
        
    # Set Chrome options to run in headless mode
    chrome_options = Options()
    chrome_options.add_argument("--headless")
        
    url = "https://www.biostadt.com" + str(productlinks[x]).lstrip("..")
    
    driver = webdriver.Chrome()
    driver.get(url)

    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content,"html.parser")
    
    name,category,tname,para,mode,crop,tar,dose,pack,dilu,beni,fea,img = None,None,None,None,None,None,None,None,None,None,None,None,None
    cropl,tarl,dosel,packl, dilul = [],[],[],[],[]
    
    
    try:
        div = soup.find("div",{"class":"tab_content"})
        name = div.find("h2").text.strip()
#         print(name)
    except:
        pass
    
    try:
        div = soup.find("div",{"class":"tab_content"})
        table = div.find("table")
        tr = table.find_all("tr")
        for i in range(len(tr)):
            td = tr[i].find_all("td")
            if "common" in td[0].text.strip().lower():
                tname = td[1].text.strip()
#         print(tname)
    except:
        pass
    
    try:
        ul = soup.find("ul",{"class":"product-pic"})
        imgl = ul.find("img").get("src")
        img = "https://www.biostadt.com" + imgl.lstrip("..")
#         print(img)
    except:
        pass
    
    
    try:
        ul = soup.find("ul",{"class":"product-pic"})
        table1 = ul.find_next_sibling("table")
        tr1 = table1.find_all("tr")
        table2 = tr1[1].find("table")
        tr2 = table2.find("tr")
        table3 = tr2.find("table")
        tr = table3.find_all("tr")
        prodtr = tr
        
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "introduction" in tx:
                para = tr[i+1].text.strip()
#         print(para)
    except:
        pass
    
    try:
        tr = prodtr
#         print(tr)
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "mode of action:" in tx:
#                 print(tx)
                mode = tr[i+1].text.strip()
#         print(mode)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "benefits:" in tx:
                print(tx)
                beni = tr[i+1].text.strip()
#         print(beni)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "features:" in tx:
                fea = tr[i+1].text.strip()
#         print(fea)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "active ingredient:" in tx:
                tname = tr[i+1].text.strip()
#         print(tname)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "method of application:" in tx:
                pack = tr[i+1].text.strip()
#         print(pack)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "packing available:" in tx or "packaging:" in tx:
                pack = tr[i+1].text.strip()
#         print(pack)
    except:
        pass
    
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "dosage & dilution:" in tx:
                
                tbody = tr[i+1].find("tbody")
                trs = tbody.find_all("tr")
                for i in range(len(trs)):
                    if i == 0:
                        continue
                    td = trs[i].find_all("td")
                    cropl.append(td[0].text.strip())
                    tarl.append(td[1].text.strip())
                    dosel.append(td[2].text.strip())
                    
        crop = " | ".join(cropl)
        tar = " | ".join(tarl)
        dose =  " | ".join(dosel)
#         print(crop)
#         print(tar)
#         print(dose)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "spectrum of action and suggested uses:" in tx:
                
                tbody = tr[i+1].find("tbody")
                trs = tbody.find_all("tr")
                for i in range(len(trs)):
                    if i == 0:
                        continue
                    td = trs[i].find_all("td")
                    cropl.append(td[0].text.strip())
                    tarl.append(td[1].text.strip())
                    dosel.append(td[2].text.strip())
                    
        crop = " | ".join(cropl)
        tar = " | ".join(tarl)
        dose =  " | ".join(dosel)
#         print(crop)
#         print(tar)
#         print(dose)
    except:
        pass
    
    try:
        tr = prodtr
        for i in range(len(tr)):
            tx = tr[i].text.strip().lower()
            if "recommendation:" in tx:
                tbody = tr[i+1].find("tbody")
                trs = tbody.find_all("tr")
                for i in range(len(trs)):
                    if i == 0 or i == 1:
                        continue
                    td = trs[i].find_all("td")
                    cropl.append(td[0].text.strip())
                    tarl.append(td[1].text.strip())
                    dosel.append(td[2].text.strip())
                    dilul.append(td[3].text.strip())
                    
                    
        crop = " | ".join(cropl)
        tar = " | ".join(tarl)
        dose =  " | ".join(dosel)
        dilu = " | ".join(dilul)
#         print(crop)
#         print(tar)
#         print(dose)
    except:
        pass
    
    

        


        
 
        
        
    
    
    print("_________end_______________" )

        
    product = {"Name":name, "Brand Name":"NACL", "Technical Name":tname,"Category":"Adjuvants","Product Brief":para, "Mode":mode,"Crop":crop,"Target":tar,"Dosage":dose,"Pack Sizes":pack,"Dilution":dilu, "Features":fea, "Benefits":beni,"Image Link":img, "Product Link": url}

    datas.append(product)

    driver.quit()
```
