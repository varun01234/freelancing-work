```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
```

```python
product_name = ["academy", "admit", "armatura", "mitigate", "suggest", "trofy", "arrest", "chariot", "cidial", "debarr", "deuce",
               "deuce-power", "dismis", "domark", "eco-champ", "eco-champ-plus", "eco-light", "fantic-m", "favor", "fenval", "fenval-dust",
               "forward", "highlight","invest", "isacarb", "isagor", "isavit", "judgement", "lacer", "lifegold", "lurit", "macario", "mangae",
               "pasport", "permit", "pocket", "polero", "precede", "proforce", "radar", "rapigro-granules", "rapigro-liquid", "raze", "ringmaster",
               "rout","salvator", "siapton-10-l", "terror-rl", "terror-super", "torpedo", "touchstone", "trofy", "ultimare", "upgrade", "valero",
               "voltage", "warlord", "warrant"]
```

```python
prodlink = []

for i in product_name:
    s = "https://www.jivagro.com/product/"+str(i)+"/"
    prodlink.append(s)
```

```python
lis = []
for i in prodlink:

    get_url = rq.get(i)
    soup = BeautifulSoup(get_url.text, "html.parser")

    try:
      img_link = soup.find("div",{"class":"woocommerce-product-gallery__image"})
      img = img_link.find("a").get('href')
    except:
      img = None

    try:
        category = soup.find("div",{"class":"main_cate"}).text
        c = category.split()
        n = len(c)
        cat = c[n-1]
        c.pop(0)
        c.pop(n-2)
        spec = " ".join(c)
    except:
        cat = None
        spec = None

    try:
        name = soup.find("h1",{"class":"product_title entry-title"}).text
    except:
        name = None

    try:    
        details = soup.find("div",{"class":"woocommerce-product-details__short-description"})
        pdetails = details.find_all("p")
        para = pdetails[1].text
    except:
        para = None


    try:
        targets = soup.find("div",{"class":"targets"})
        ptargets = targets.find_all('p')
        pe = ptargets[0].text
        p = pe.split()
        p.pop(0)
        p.pop(0)
        pest = " ".join(p)
    except:
        pest = None

    try:
        targets = soup.find("div",{"class":"targets"})
        ptargets = targets.find_all('p')
        pe = ptargets[1].text
        p = pe.split()
        p.pop(0)
        p.pop(0)
        dos = " ".join(p)
    except:
        dos = None

    try:
        targets = soup.find("div",{"class":"targets"})
        ptargets = targets.find_all('p')
        pe = ptargets[2].text
        p = pe.split()
        p.pop(0)
        p.pop(0)
        crp = " ".join(p)
    except:
        crp = None


    try:
        m = soup.find("div",{"class":"panel-body"})
        fe = m.find("p").text
    except:
        fe = None


    try:
        m = soup.find_all("div",{"class":"panel panel-default"})
        k = m[2].find("h4",{"class":"panel-title"}).text

        ks = k.strip()

        if ks == "Packaging":
            ml = m[2].find("li").text
        else:
            ml = None

    except:
        ml = None


    data = {"Product Name" : name,
            "Brand Name" : "Jivagro",
            "Category": cat,
            "Specification": spec,
            "Product Brief": para,
            "Crop": crp,
            "Targest Pest": pest,
            "Dosage":dos,
            "Feature": fe,
            "Product Image": img,
            "Packaging": ml,
            "Product Link": i
            }

    lis.append(data)

```

```python
new_data = pd.DataFrame.from_dict(lis)
new_data.head(60)
```

```python

# saving the dataframe
new_data.to_csv('jivagro.csv')
```
