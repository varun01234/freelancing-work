```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
```

```python
driver = webdriver.Chrome()
url= "https://ariesagro.com/products-services/"
driver.get(url)


wait = WebDriverWait(driver, 10) # wait for up to 10 seconds for the page to load
wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

# try:
#     label_element = driver.find_element(By.XPATH, '//label[text()="All"]')
#     label_element.click()
#     print("clicked")
# except:
#     pass

# tm.sleep(2)
content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")

links = soup.find_all("div",{"data-widget_type":"image.default"})

# ul = soup.find("ul",{"class":"breadcrumbs"})
# li = ul.find_all("li")
# cat = li[len(li)-1].text.strip()
# print(cat)

productlinks = []
for i in links:
    try:
        a = i.find("a")
        link = a.get("href")
        productlinks.append(link)
    except:
        pass

print(len(productlinks))


driver.quit()
```

```python
for i in  productlinks:
    print(i)
```

```python
datas = []
for x in range(1, len(productlinks)):


#     if x == 2:
#         break
        
    # Set Chrome options to run in headless mode
    chrome_options = Options()
    chrome_options.add_argument("--headless") 
        
    url =  str(productlinks[x])
    
    driver = webdriver.Chrome()
    driver.get(url)

    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content,"html.parser")
    
    name,cat,tname,para,mode,crop,tar,dose,pack,dilu,beni,fea,img = None,None,None,None,None,None,None,None,None,None,None,None,None
    cropl,tarl,dosel,packl, dilul = [],[],[],[],[]
    
    
    try:
        div = soup.find("div",{"data-widget_type":"heading.default"})
        cat = div.find("h2").text.strip()
#         print(cat)
    except:
        pass
    
    try:
        names = soup.find_all("h2",{"class":"elementor-heading-title elementor-size-default"})
        name = names[1].text.strip()
#         print(name[1].text.strip())
    except:
        pass
    
    
    
    try:
        div = soup.find("div",{"data-widget_type":"accordion.default"})
#         print(div)
        items = div.find_all("div",{"class":"elementor-accordion-item"})
        for i in items:
            tx = i.find("a").text.strip()
            print(tx)
            if "Product features" in tx:
                ul = i.find("ul").text.strip()
                fea = ul
            elif "Contents" in tx:
                ul = i.find("ul").text.strip()
                tname = ul
            elif "Pack Size" in tx:
                divs = i.find_all("div")
                pack = divs[1].text.strip()
                
            elif "Dosage" in tx:
                ul = i.find("ul").text.strip()
                dose = ul
    except:
        pass
    
    try:
        beni = soup.find("div",{"class":"elementor-tabs"}).text.strip().lstrip("Benifits \n")
#         print(beni)
    except:
        pass

    try:
        img = soup.find("img",{"width": "380"}).get("src")
#         print(img)
    except:
        img = None
    

        


        
 
        
        
    
    
    print("_________end_______________", x )

        
    product = {"Name":name, "Brand Name":"ARIES", "Technical Name":tname,"Category":cat,"Product Brief":para, "Mode":mode,"Crop":crop,"Target":tar,"Dosage":dose,"Pack Sizes":pack, "Features":fea, "Benefits":beni, "Image Link":img, "Product Link": url}

    datas.append(product)

    driver.quit()
```

```python
new_data = pd.DataFrame.from_dict(datas)
pd.set_option('display.max_rows', None)
new_data.head(110)
```
