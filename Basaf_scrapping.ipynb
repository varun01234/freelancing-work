```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
```

```python
driver = webdriver.Chrome()
url= "https://www.unimart.in/products/upl/"
# driver.maximize_window()
driver.get(url)

tm.sleep(5)
content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")
link = soup.findAll("div",{"class":"col-sm-6 col-md-6 col-lg-3"})
links = link[0].find("a").get('href')
print(links)
productlinks = []
for i in range(len(link)):
        links = link[i].find("a").get('href')
        productlinks.append(links)
# print(productlinks)
print(len(productlinks))

driver.quit()
```

```python
data = []
for x in range(0, 108):
    url = "https://www.unimart.in/products/upl/" + str(productlinks[x]) + '/'
    
    driver = webdriver.Chrome()
    # driver.maximize_window()
    driver.get(url)

    tm.sleep(5)
    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content,"html.parser")
    
    try:
        name = soup.find("h1",{"class":"page-titles"}).text
    except:
        name = None
    
    try:
        category = productlinks[x]
        csp = category.split("/")
        cat = csp[0]
    except:
        cat = None

    try:
        tabl = soup.find("tbody",{"class":"table_body_bg"})
        crop = tabl.find_all("td")
        b=[]
        for i in range(len(crop)):
            if i%2==0:
                x = crop[i].text
                b.append(x)
        crp = " | ".join(b)
    except:
        crp = None
        
    try:
        a=[]
        targ = soup.find_all("span",{"class":"target-name"})
        for i in range(len(targ)):
            x = targ[i].text
            a.append(x)
        tar = " | ".join(a)
    except:
        tar = None

    try:
        para = soup.find("div",{"class":"right-pbox"})
        paran = para.find_all("p")
        pli=[]
        for i in paran:
            pli.append(i.text)
        plia = " | ".join(pli).strip(" | ")
    except:
        plia = None
    
    try:
        feat = soup.find_all("div",{"class":"right-pbox"})
        fe = feat[1].find_all("p")
        feli=[]
        for i in fe:
            feli.append(i.text)
        feans = " | ".join(feli)
    except:
        feans = None
        
    try:
        img = soup.find("img",{"class":"img-responsive xzoom lazyload"}).get('src')
    except:
        img = None
        
    try:
        speci = soup.find_all("div",{"class":"product-details-text mb-4 border-bottom"})
        model = []
        for i in range(len(speci)):
            mode = speci[i].find("h2",{"class":"pbox-title"}).text
            if mode.strip() == "Mode Of Action":
                modep = speci[i].find_all("p")
                for i in range(len(modep)):
                    x = modep[i].text
                    model.append(x)
        spec = "".join(model)
    except:
        spec = None
        
    d=""

    try:    
        doses = soup.find_all("div",{"class":"product-details-text mb-4 border-bottom"})
        dosesl = []
        for i in range(len(doses)):
            dose = doses[i].find("h2",{"class":"pbox-title"}).text
            if dose.strip() == "Dose / Pack Size" or dose.strip() == "Dose/pack Size":
                doserow = doses[i].find_all("tr")
                for i in range(1,len(doserow)):

                    dosecol = doserow[i].find_all("td")
                    for i in range(len(dosecol)):
                        if i == (len(dosecol)-1):
                            x = dosecol[i].text
                            dosesl.append(x)
        d = " | ".join(dosesl)

    except:
        d=""

    if d == "":
        try:
            doses = soup.find_all("div",{"class":"product-details-text mb-4 border-bottom"})
            dosesl = []
            for i in range(len(doses)):
                dose = doses[i].find("h2",{"class":"pbox-title"}).text
                if dose.strip() == "Dose / Pack Size" or dose.strip() == "Dose/pack Size":
                    dosep = doses[i].find_all("p")
                    for i in range(len(dosep)):
                        x = dosep[i].text
                        dosesl.append(x)
            d = " | ".join(dosesl).strip(" | ")
        except:
            d=None
                    
        
    try:
        pack = soup.find_all("div",{"class":"product-details-text mb-4 border-bottom"})
        packl =  []
        for i in range(len(pack)):
            pac = pack[i].find("h2",{"class":"pbox-title"}).text
            if pac.strip() == "Sku Available":
                pacp = pack[i].find_all("p")
                for i in range(len(pacp)):
                    x = pacp[i].text
                    packl.append(x)
        ml = "".join(packl)
    except:
        ml = None
        
    datadict = {"Product Name" : name,
            "Brand Name" : "unimart",
            "Category": cat,
            "Specification": spec,
            "Product Brief": plia,
            "Crop": crp,
            "Targest Pest": tar,
            "Dosage":d,
            "Feature": feans,
            "Product Image": img,
            "Packaging": ml,
            "Product Link": url
            }
    data.append(datadict)


    driver.quit()
```

```python
new_data = pd.DataFrame.from_dict(data)
pd.set_option('display.max_rows', None)
new_data.head(109)
```
