```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python

driver = webdriver.Chrome()
url= "https://crop-protection.basf.in/en/pgrmicronutrients"
driver.get(url)

# tm.sleep(5)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
tm.sleep(5)

SCROLL_PAUSE_TIME = 0.5 
    # Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
        # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
    tm.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")


div = soup.find_all("div",{"class":"views-row"})
productlinks = []
for i in div:
    link = i.find("a").get("href")
    productlinks.append(link)  


print(len(productlinks))

driver.quit()
```

```python
datas = []
for x in range(0, len(productlinks)):


#     if x == 3:
#         break
        
    url = "https://crop-protection.basf.in" + str(productlinks[x])
    
    driver = webdriver.Chrome()
    driver.get(url)
    
#     wait = WebDriverWait(driver, 10) # wait for up to 10 seconds for the page to load
#     wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

#     for i in range(30):
#         driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.DOWN)
#         tm.sleep(0.1)


    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content,"html.parser")


    try:
        name = soup.find("h1",{"class":"page-title"}).text.strip()
#         print(name)
    except:
        name=None
        
    
    try:
        imgs = soup.find("img",{"data-entity-type":"file"}).get('src')
        img = "https://crop-protection.basf.in" + imgs
#         print(img)
    except:
        img=None
        
    try:
        pdiv = soup.find("div",{"class":"layout__region layout__region--content"})
        div = pdiv.find("div",{"class":"clearfix text-formatted field field--name-body field--type-text-with-summary field--label-hidden field__item"})
        fea = div.find('p').text
#         print(appl)
    except:
        fea = None
        
        
    
    try:
        pdiv = soup.find("div",{"class":"layout__region layout__region--content"})
        div = pdiv.find("div",{"class":"clearfix text-formatted field field--name-body field--type-text-with-summary field--label-hidden field__item"})
        para = div.find('p').text
#         print(para)
    except:
        para = None

    try:
        pdiv = soup.find("div",{"class":"layout__region layout__region--content"})
        div = pdiv.find("div",{"class":"field field--name-field-components field--type-entity-reference-revisions field--label-hidden"})
        bediv = div.find("div",{"class":"downloads-box grid grid-twocol-six paragraph paragraph--type--benefits-box paragraph--view-mode--default"})
        beni = div.find('ul').text
#         print(beni)
    except:
        beni = None

    
    
    try:
        cropl = []
        pdiv = soup.find("div",{"class":"layout__region layout__region--content"})
        sdiv = pdiv.find("div",{"class":"field field--name-field-components field--type-entity-reference-revisions field--label-hidden"})
        div = sdiv.find_all("div",{"class":"clearfix text-formatted field field--name-field-body field--type-text-long field--label-hidden field__item"})

        for i in div:

            head = i.find("h3")
#             print(head)
            if head != None:
                head = head.text.strip().lower()
                if "application" in head:
#                     print(head)
                    dl = i.find("dl")
                    dt = dl.find_all("dt")
                    for t in dt:
                        cropl.append(t.text.strip())
#                         print(t.text)
                        head = i.find("h3")
#             print(head)
            head2 = i.find("h2")
            if head2 != None:
                head2 = head2.text.strip().lower()
                if "application" in head2:
#                     print(head2)
                    dl = i.find("dl")
                    dt = dl.find_all("dt")
                    for t in dt:
                        cropl.append(t.text.strip())
#                         print(t.text)

        crop = " | ".join(cropl)
#         print(crop)             
    except:
        crop = None
      
    
    dosel, targetl, usel = [],[],[]
    try:
        
        pdiv = soup.find("div",{"class":"layout__region layout__region--content"})
        sdiv = pdiv.find("div",{"class":"field field--name-field-components field--type-entity-reference-revisions field--label-hidden"})
        div = sdiv.find_all("div",{"class":"clearfix text-formatted field field--name-field-body field--type-text-long field--label-hidden field__item"})
        for i in div:

            head = i.find("h3")
            if head != None:
                head = head.text.strip().lower()
                if "application" in head:
#                     print(head)
                    dl = i.find("dl")
                    dd = dl.find_all("dd")
        
                    for t in dd:
                        td = t.find_all("td")
                        th = t.find_all("th")
                        for j in range(len(th)):
                            use = th[j].text.strip().lower()
                            if "apply" in use:
                                n=j
                        dosel.append(td[1].text.strip())
                        targetl.append(td[0].text.strip())
                        usel.append(td[n].text.strip())
#                         print(td[0].text.strip())
#                         print(td[1].text.strip())
#                         print(td[n].text.strip())
                        
            head2 = i.find("h2")
            if head2 != None:
                head2 = head2.text.strip().lower()
                if "application" in head2:
                    dl = i.find("dl")
                    dd = dl.find_all("dd")
                    
                    
                    for t in dd:
                        td = t.find_all("td")
                        th = t.find_all("th")
                        for j in range(len(th)):
                            use = th[j].text.strip().lower()
                            if "apply" in use:
                                n=j
                                
                        dosel.append(td[1].text.strip())
                        targetl.append(td[0].text.strip())
                        usel.append(td[n].text.strip())
#                         print(td[0].text.strip())
#                         print(td[1].text.strip())
#                         print(td[n].text.strip())
            
    except:
        pass
    
    dose = " | ".join(dosel)
    target = " | ".join(targetl)
    apply = " | ".join(usel)
         
        
    tech,mode,pack = None,None,None
    
    
    print("_________end_______________" , x)

        
    product = {"Name":name, "Brand Name":"BASF", "Technical Name":tech,"Category":"Micronutrients","Product Brief":para, "Mode":mode,"Crop":crop,"Target":target,"Pack Sizes":pack,"Dosage":dose, "Features":fea, "Benefits":beni, "How to use": apply ,"Image Link":img, "Product Link": url}

    datas.append(product)

    driver.quit()
```
