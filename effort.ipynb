```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
import time as tm
data = []
productlinks = ["https://www.coromandel.biz/product-service/groshakti-plus/",                "https://www.coromandel.biz/product-service/groshakti/",                "https://www.coromandel.biz/product-service/grosmart/",                "https://www.coromandel.biz/product-service/paramfos/",                "https://www.coromandel.biz/product-service/gromor-28-28-0/",                "https://www.coromandel.biz/product-service/gromor-20-20-0-13/",                "https://www.coromandel.biz/product-service/gromor-15151509/",                "https://www.coromandel.biz/product-service/gromor-123216/",                "https://www.coromandel.biz/product-service/gromor-ultra-102626/",                "https://www.coromandel.biz/product-service/gromor-10-26-26/",                "https://www.coromandel.biz/product-services/comple",                "https://www.coromandel.biz/product-services/dap/",                "https://www.coromandel.biz/product-services/mop/",                "https://www.coromandel.biz/product-services/ssp/",                "https://www.coromandel.biz/product-services/urea/"]

for x in range(0, 15):
    driver = webdriver.Chrome()
    driver.get("https://www.coromandel.biz/product-services/complex/")

    url = str(productlinks[x])

    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content,"html.parser")

    try:
        name = soup.find("h4",{"class":"yellow-title text-left pb-3"}).text
    except:
        name = None
    
    try:
        pack = soup.find("strong",{"class":"pack-size"}).text
    except:
        pack = None
    
    try:
        briefs = soup.find("div",{"class":"product-details pt-4"})
        bps = briefs.find_all("p")
        bpsl=[]
        for i in bps:
            bpsl.append(i.text)
        det = " ".join(bpsl)
            
    except:
        det = None
    
    try:
        do=""
        panel = soup.find_all("div",{"class":"panel panel-default"})
        for i in panel:
            head = i.find("h4").text.strip()
            if head == "Dosage and Application" or head == "Dosage & Time of Application" or head == "Dosage & Application" or head == "Dosage":
                paras = i.find_all("li")
                if len(paras)==0:
                    paras = i.find_all("p")
                dosel=[]
                for i in paras:
                    dosel.append(i.text)
                do = " ".join(dosel)
    except:
        do = None

    try:
        fe=""
        panel = soup.find("div",{"class":"panel panel-default active"})
        head = panel.find("h4").text.strip()
        li = panel.find_all("li")
        featurel=[]
        for i in li:
            featurel.append(i.text)
        fe = "\n".join(featurel)
    except:
        fe = None
        
    try:
        img = soup.find("img", {"class":"w-100"}).get('src')
    except:
        img = None
        
    try:
        category = soup.find_all("div", {"class":"sub-title"})[1].text
    except:
        category = None
        
    try:
        benefit = soup.find_all("div", {"class":"sub-title"})[2].text
    except:
        benefit = None

    product = {"Product Name": name, "Brand Name": "Coromandel", "Detail": det, "Dose": do, "Pack Size": pack, "Features": fe,"Image Link": img, "Product Link": url, "Category": category, "Benefit": benefit}
    data.append(product)
    print(x)
    
    driver.quit()
```
